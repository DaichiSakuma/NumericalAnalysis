```python
import matplotlib.pyplot as plt
import japanize_matplotlib
import numpy as np

##m*(d^2x/dt^2)+c*(dx/dt)+kx=0の

def f1(x, v):
    return v
def f2(x, v):
    return -2*x-0.2*v

h =5
for i in range(1, 5):
    h = h/5 #差分
    list_x = []
    list_vx = []
    list_t = []
    t = 0
    #初期条件
    k = 2
    c = 0.2
    m = 1
    xi = 10
    vxi = 0
    while t<=30:
        t+=h
        list_x.append(xi)
        list_vx.append(vxi)
        list_t.append(t)
        tempxi = xi
        tempvxi = vxi
        xi += h*f1(tempxi, tempvxi)
        vxi += h*f2(tempxi, tempvxi)
    plt.plot(list_t, list_x, label='Euler')
    t = np.arange(0,30, 0.01)
    # 2step:上のリストに対して関数を適用
    x =10*(np.exp(-0.1*t))*(np.cos((1.99**0.5)*t))
    # 3step:plt.plot(x軸を指定, y軸を指定)
    plt.plot(t, x, label='解析解')
    plt.title('               Newton法における、h='+str(h)+'のときの減衰振動の様子')
    plt.xlabel('t')
    plt.ylabel('x')
    plt.rcParams['axes.grid'] = True
    name = 'N-'+str(i)+'.png'
    plt.legend()
    plt.savefig(name)
    plt.show()


h = 25

for i in range(1, 5):
    k = 2
    c = 0.2
    m = 1
    xi = 10
    vxi = 0
    h = h/5
    list_x = []
    list_vx = []
    list_t = []
    loop = int(30/h)
    for t in range(loop):
        kx1 = h*f1(xi, vxi)
        kvx1 = h*f2(xi, vxi)
        kx2 = h*f1(xi+(2.0/3.0)*kx1, vxi+(2.0/3.0)*kvx1)
        kvx2 = h*f2(xi+(2.0/3.0)*kx1, vxi+(2.0/3.0)*kvx1)
        kx3 = h*f1(xi+(2.0/3.0)*kx2, vxi+(2.0/3.0)*kvx2)
        kvx3 = h*f2(xi+(2.0/3.0)*kx2, vxi+(2.0/3.0)*kvx2)
        list_x.append(xi)
        list_vx.append(vxi)
        list_t.append(t*h)
        xi += (2*kx1+3*kx2+3*kx3)/8
        vxi += (2*kvx1+3*kvx2+3*kvx3)/8
    t = np.arange(0,30, 0.01)
    # 2step:上のリストに対して関数を適用
    x =10*(np.exp(-0.1*t))*(np.cos((1.99**0.5)*t))
    # 3step:plt.plot(x軸を指定, y軸を指定)
    plt.plot(t, x, label='解析解')
    plt.plot(list_t, list_x, label='RK3')
    plt.title('               3次Runge-Kutta 法における、h='+str(h)+'のときの減衰振動の様子')
    plt.xlabel('t')
    plt.ylabel('x')
    plt.legend()
    name = 'RK-3-'+str(i)+'.png'
    plt.savefig(name)
    plt.show()


h = 25
for i in range(1, 5):
    k = 2
    c = 0.2
    m = 1
    xi = 10
    vxi = 0
    h = h/5
    list_x = []
    list_vx = []
    list_t = []
    loop = int(30/h)
    for t in range(loop):
        kx1 = h*f1(xi, vxi)
        kvx1 = h*f2(xi, vxi)
        kx2 = h*f1(xi+0.5*kx1, vxi+0.5*kvx1)
        kvx2 = h*f2(xi+0.5*kx1, vxi+0.5*kvx1)
        kx3 = h*f1(xi+0.5*kx2, vxi+0.5*kvx2)
        kvx3 = h*f2(xi+0.5*kx2, vxi+0.5*kvx2)
        kx4 = h*f1(xi+kx3, vxi+kvx3)
        kvx4 = h*f2(xi+kx3, vxi+kvx3)
        list_x.append(xi)
        list_vx.append(vxi)
        list_t.append(t*h)
        xi += (kx1+2*kx2+2*kx3+kx4)/6
        vxi += (kvx1+2*kvx2+2*kvx3+kvx4)/6
    t = np.arange(0,30, 0.01)
    # 2step:上のリストに対して関数を適用
    x =10*(np.exp(-0.1*t))*(np.cos((1.99**0.5)*t))
    # 3step:plt.plot(x軸を指定, y軸を指定)
    plt.plot(t, x, label='解析解')
    plt.plot(list_t, list_x, label='RK4' )
    plt.title('                  4次Runge-Kutta 法における、h='+str(h)+'のときの減衰振動の様子')
    plt.xlabel('t')
    plt.ylabel('x')
    plt.legend()
    name = 'RK4-'+str(i)+'.png'
    plt.savefig(name)
    plt.show()
```
